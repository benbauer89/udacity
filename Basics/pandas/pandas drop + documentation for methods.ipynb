```python
drinks.rename(columns= {'continent': 'Continent'}, inplace=True) # inplace true affect the underlying df
```

```python
## select based on condition (subsetting)
drinks[drinks.beer_servings >= 250]
```

```python
# same condition but select only specific columns of that dataframe
drinks[drinks.beer_servings >= 250][['country', 'beer_servings']]
```

```python
# multile conditions connected via and
drinks[(drinks.beer_servings >= 250) & (drinks.wine_servings >= 100)][['country', 'beer_servings', 'wine_servings']]
```

```python
# multile conditions connected via or
drinks[(drinks.beer_servings >= 250) | (drinks.wine_servings >= 100)][['country', 'beer_servings', 'wine_servings']]
```

```python
# can also pass a list of arguments to isin()
drinks[(drinks.beer_servings >= 250) & (drinks.wine_servings >= 100) & drinks.Continent.isin(['Europe'])]
```
